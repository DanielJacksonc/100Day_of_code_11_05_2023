```python
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}



def check_resources(user_res):
    for item in user_res:
        if user_res[item] >= resources[item]:
            print("Sorry there is not enough water")
            return False
        else:
            return True

def process_coin():
    print("Please insert coin. ")
    money = int(input("Please provide quaters "))*0.25
    money += int(input("Please provide dimes "))*0.10
    money += int(input("Please provide nickels "))*0.05
    money += int(input("Please provide pennies "))*0.01
    return money
def transaction(user_money, drink_cost):
    if user_money >= drink_cost:
        global profit
        change = round(user_money - drink_cost, 2)
        profit += drink_cost
        print(f"your money is accepted and your change is ${change}")
        return True
    else:
        print("Sorry that's not enough money. Money refunded.")
        return False
    
        
def report(drink_reso):
    for item in drink_reso:
        resources[item] -= drink_reso[item]
        
    print("Here is your latte. Enjoy!ðŸ¥«")
    return True


profit = 0
machine_on = True
while machine_on:
    
    user_prompt = input("What would you like? espresso/latte/cappuccino or off, report:").lower()
    if user_prompt == "off":
        print("You hae successfully turned machine OFF. Bye!ðŸ˜Š!")
        machine_on = False
    elif user_prompt == "report":
        print(f"Water : {resources['water']}")
        print(f"Coffee : {resources['coffee']}")
        print(f"Milk : {resources['milk']}")
        print(f"Profit : {profit}")
    else:
        drink = MENU[user_prompt]
        drink_res = drink["ingredients"]
        drink_price = drink['cost']
        if check_resources(drink_res):
            pay = process_coin()
            if transaction(pay, drink_price):
                report(drink_res)
                
            
        



    
        
```
